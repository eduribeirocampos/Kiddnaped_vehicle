```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.lines as lines
import numpy as np

path = ('../data/')

with open(path+'map_data.txt', 'r') as infile, open(path+'outfile.csv','w') as outfile: 
    for line in infile: 
        outfile.write(line.replace('\t',','))

        
df = pd.read_csv(path+'outfile.csv',delimiter=',',names=['x','y','landmark_id']) 
df.head()
```

```python
fig, ax = plt.subplots(figsize=(15,8))
ax.scatter(df.x,df.y,s=550,marker='.', color='blue',alpha=0.2)

ax.set_title('map_data - Visualization', fontsize=25)
ax.set_xlabel('X Coordinate', fontsize=20)
ax.set_ylabel('Y Coordinate', fontsize=20)
ax.axvline(linewidth=2, color='r')
ax.axhline(linewidth=2, color='r')
ax.grid(True)
#ax.axis('equal')
ax.set_aspect('equal', 'box')
ax.set(xlim=(-50,300), ylim=(-125,50))


# Major ticks every 20, minor ticks every 5
major_ticksx = np.arange(-50, 301, 50)
minor_ticksx = np.arange(-50, 301, 25)
major_ticksy = np.arange(-100, 50, 50)
minor_ticksy = np.arange(-100, 50, 25)

ax.set_xticks(major_ticksx)
ax.set_xticks(minor_ticksx, minor=True)
ax.set_yticks(major_ticksy)
ax.set_yticks(minor_ticksy, minor=True)

# Or if you want different settings for the grids:
ax.grid(which='minor', alpha=0.2)
ax.grid(which='major', alpha=0.5)



for i, txt in enumerate(df.landmark_id):
    ax.annotate(txt, (df.x[i], df.y[i]),fontsize=12,ha='center',va='center')


#plt.savefig('map_data_visualization.jpg')
plt.show()
```
